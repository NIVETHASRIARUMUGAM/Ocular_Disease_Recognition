```python
import matplotlib.pyplot as plt

# Names of neural network architectures
models = ["EfficientNet", "DenseNet", "ResNet", "VGG19", "AlexNet", "CNN"]

# Corresponding accuracy values
accuracy = [30, 90, 60, 83, 73, 79]

# Create a bar plot
plt.figure(figsize=(10, 6))  # Set the figure size

plt.bar(models, accuracy, color=['lightblue', 'orange', 'green', 'red', 'purple', 'brown'])

# Add labels and title
plt.xlabel("Model Architecture")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy of Different Neural Network Architectures")

# Display the plot
plt.show()  # If you're in a Jupyter Notebook, this will display the plot in the output
```
